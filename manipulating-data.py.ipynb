# Using Data in Pandas

Using pandas--a brief guide to learning data-visualisation. Learning to work with data. 

```python
import pandas as pd
# dataset: AppleStore (data of applications on apple-store)
dataframe = pd.read_csv('./AppleStore.csv')
dataframe.head(3)
```

```python
df = dataframe
# We use `df` as shortname for dataframe. 
df.iloc[2]
```

- Use `.iloc` to access a row, most simply; Use it as - `dataframe.iloc[row-index]` (from as usual, first row as zero, just like list indices).

- Method `.loc` can be used for passing a value from your dataframe's index (some unique-id) as `dataframe.loc[unique-id]`

```python
df.iloc[[2, 1, 0]]
```

**Slicing the dataframe works as if it were a list.**

```python
df[:3]
```

```python
df[2:7]
```

```python
df['track_name'].head(3)

# df.track_name.head(3) works the same.
```

We can access a single column as if it were an attribute, but only if the name has no spaces, uses only basic characters, and doesn't share a name with a dataframe method--too much! Duhh!. **'Python-Keep-In-Mind'**: Use names (column names--here) with proper name, and use '_' in place of ' '.

**A nice cleaning function for such tidier situation.**

```python
df.columns = [col.replace(' ', '_').lower() for col in df.columns]
print(df.columns)
```

```python
df[['price', 'size_bytes']]
```

Above queries is good when need to stack up multiple column names.

- Passing a list will give different return type -- with unique id, `Name` and `dtype`.

```python
df.track_name.iloc[2]
```

```python
df.track_name.iloc[[2]]
```

### Selecting a Subset of the Data

```python
(df.track_name == "WeatherBug - Local Weather, Radar, Maps, Alerts").head(3)
```

```python
df[df.track_name == "WeatherBug - Local Weather, Radar, Maps, Alerts"]
```

We can use complicated logical expressions -- including lambda. For example,

```python
df.columns
```

```python
df[(df.user_rating > 4.5) | (df.price > 10.0)].head(3)
```

We can use `.isin` to get all values equal to a given value.

```python
df[df.user_rating.isin(['5.0'])].head()
```

## Selecting and Filtering Data

Datasets have too many variables to wrap around. Selecting and filtering helps. Let's work with functions, and how to do that.

```python
import pandas as pd

filepath = './AppleStore.csv'
df = pd.read_csv(filepath) # DataFrame
print(df.columns)
```

### Select Single Column

Use **dot-notation** to pull out a variable or column -- as *dataframe.column_name*.

**NOTE**: Single column of data is stored in Series, which is broadly like a DataFrame with only a single column of data.

```python
df.price.head()
```

### Select Multiple Column

By providing a list of column names. Like -

```python
two_col_data = df[['price', 'user_rating']]
```

`describe()` method calculates basic count, mean, std, ... and much more.

```python
two_col_data.describe()
```
